# Data Preprocessing -  Data Cleaning

```python
data.head(3)
```

# ------------- Transform the values of X.ID column and Response column ---------


```python
#------------------------ Transform the values of X.ID column and Response column ---------------#


#--- 1- for X.ID
myfun1 = map(lambda s: s[1:], data['X.ID'])

newID = pd.Series(list(myfun1))
data['X.ID'] = newID

#--- 2- for Response
myfun2 = map(lambda s: s[:1], data['Response'])

newResponse = pd.Series(list(myfun2))
data['Response'] = newResponse

```

```python
data.head()   #--------------------------------- Done ---------------------------------#
```

#     ------ Change the Data types to convenient format - Data Transofmation -----


```python

#-------------------  1- Change columns from object type to numerical data ----------------------#

data[['X.ID','Response']]= data[['X.ID','Response']].apply(pd.to_numeric,errors='coerce')

```

```python
data.info()  #------------------------------ 1 - Done -------------------------------#
```

```python
#-------------------  2- Change Dt_Customer from object type to Data data type ----------------------#

data['Dt_Customer'] = pd.to_datetime(data["Dt_Customer"], infer_datetime_format=True, dayfirst=True)
```
